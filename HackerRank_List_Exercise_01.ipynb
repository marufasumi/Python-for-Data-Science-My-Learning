```python
# Read the number of commands
n = int(input())

# Initialize an empty list
lst = []

# Iterate through each command
for _ in range(n):
    # Read the command and split it into parts
    command = input().split()

    # Determine the command type and execute the corresponding operation
    ## insert() is a in-built function
    if command[0] == "insert":
        index = int(command[1])
        value = int(command[2])
        lst.insert(index, value)
    
    elif command[0] == "print":
        print(lst)  # Output the current list
    
    elif command[0] == "remove":
        value = int(command[1])
        lst.remove(value)  # Remove the first occurrence of the specified value
    
    elif command[0] == "append":
        value = int(command[1])
        lst.append(value)  # Append the specified value to the end
    
    elif command[0] == "sort":
        lst.sort()  # Sort the list in ascending order
    
    elif command[0] == "pop":
        lst.pop()  # Remove the last element
    
    elif command[0] == "reverse":
        lst.reverse()  # Reverse the list
```
